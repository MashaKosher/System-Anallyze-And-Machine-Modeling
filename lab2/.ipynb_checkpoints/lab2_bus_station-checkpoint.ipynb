# Лабораторная работа: Моделирование «Билетные кассы на автовокзале»

Цель: смоделировать работу 4 касс с отдельными очередями, учетом технологических перерывов и выбором пассажира «приблизительно наименьшей» очереди. Реализован дискретно-событийный движок без внешних библиотек.

Ключевые элементы:
- Пассажиры приезжают по пуассоновскому потоку (экспоненциальные межприходы)
- Время обслуживания — экспоненциальное
- 4 кассы: циклы работы по 60 мин (45 мин работа + 15 мин перерыв), за `T_announce` минут до перерыва — объявление и возможная миграция пассажиров
- Выбор очереди — «приблизительно минимальная» (погрешность по длине ±delta)

Выходные показатели:
- Среднее время ожидания в очереди
- Среднее время пребывания в системе
- Средняя длина очереди по каждой кассе (временная средняя)
- Кол-во пассажиров, не успевших обслужиться до перерыва и ждавших его окончания
- Общее число обслуженных пассажиров
- Кол-во ушедших из системы из-за превышения лимита ожидания (необязательно)


```python
from __future__ import annotations
from dataclasses import dataclass, field
from typing import Callable, List, Optional, Dict, Tuple, Any
import heapq
import random
import math

# -------------------------
# Глобальные параметры
# -------------------------
SEED = 42
random.seed(SEED)

MINUTES = 1.0
HOUR = 60.0 * MINUTES

PARAMS = {
    # Поток прибытия пассажиров (интенсивность лямбда, пассажиров в минуту)
    "arrival_rate": 1.0,  # λ
    # Скорость обслуживания одной кассой (μ, пассажиров в минуту)
    "service_rate_per_booth": 0.35,  # μ
    # Кол-во касс
    "num_booths": 4,
    # Цикл работы кассы: 45 мин работа, 15 мин перерыв
    "work_minutes": 45.0,
    "break_minutes": 15.0,
    # За сколько минут до перерыва оповещать
    "announce_minutes": 5.0,
    # Длительность моделирования (в минутах)
    "simulation_duration": 8 * HOUR,
    # Параметры выбора «приблизительно минимальной» очереди
    "approx_delta": 1,       # допустимая погрешность по длине очереди (шт.)
    "epsilon_random": 0.05,  # шанс выбрать не оптимальную очередь
    # Во время анонса пассажиры мигрируют с вероятностью:
    "announce_migrate_prob": 1.0,
    # Кара за "announce" для выбора (искусственно увеличиваем воспринимаемую длину)
    "announce_penalty": 2,
    # Отказ из-за ожидания (None отключает)
    "max_wait_time": None,  # в минутах; например, 60.0
}
```

```python
@dataclass
class Passenger:
    id: int
    arrival_time: float
    chosen_booth_id: Optional[int] = None
    service_start_time: Optional[float] = None
    service_end_time: Optional[float] = None
    abandoned: bool = False  # ушёл из-за лимита ожидания
    waited_over_break: bool = False  # застал перерыв и дождался

@dataclass(order=True)
class Event:
    time: float
    priority: int
    kind: str
    payload: Any = field(compare=False, default=None)

class DES:
    def __init__(self):
        self.t: float = 0.0
        self.queue: List[Event] = []
        self.event_counter: int = 0

    def schedule(self, time: float, kind: str, payload: Any = None, priority: int = 0):
        self.event_counter += 1
        evt = Event(time=time, priority=priority, kind=kind, payload=payload)
        heapq.heappush(self.queue, evt)

    def run(self, until: float, on_event: Callable[[Event, float], None]):
        while self.queue and self.queue[0].time <= until:
            evt = heapq.heappop(self.queue)
            self.t = evt.time
            on_event(evt, self.t)
        # Продвинуть время до until
        self.t = until
```

```python
class TicketBooth:
    def __init__(self, booth_id: int, params: Dict[str, Any]):
        self.booth_id = booth_id
        self.params = params
        self.queue: List[Passenger] = []
        self.is_on_break: bool = False
        self.is_in_announce: bool = False
        self.cycle_start_time: float = 0.0  # начало текущего цикла работы
        self.currently_serving: Optional[Passenger] = None
        self.next_cycle_break_start: float = 0.0
        self.next_cycle_break_end: float = 0.0
        self.next_announce_time: float = 0.0

    def perceived_queue_length(self) -> int:
        # Во время анонса добавляем штраф к воспринимаемой длине
        penalty = self.params["announce_penalty"] if self.is_in_announce else 0
        # Если касса в перерыве, считаем очередь как очень длинную
        if self.is_on_break:
            return 10**9
        return len(self.queue) + penalty

    def plan_cycle(self, now: float) -> Tuple[float, float, float]:
        work = self.params["work_minutes"]
        brk = self.params["break_minutes"]
        announce = self.params["announce_minutes"]
        # Цикл начинается в момент now, планируем ближайшие времена
        start = now
        break_start = start + work
        break_end = break_start + brk
        announce_time = max(start, break_start - announce)
        self.cycle_start_time = start
        self.next_cycle_break_start = break_start
        self.next_cycle_break_end = break_end
        self.next_announce_time = announce_time
        return announce_time, break_start, break_end

    def on_announce(self):
        self.is_in_announce = True

    def on_break_start(self):
        self.is_on_break = True
        self.is_in_announce = False

    def on_break_end(self, new_cycle_start: float):
        self.is_on_break = False
        self.is_in_announce = False
        self.cycle_start_time = new_cycle_start

class Metrics:
    def __init__(self, num_booths: int):
        self.num_booths = num_booths
        self.total_wait_time: float = 0.0
        self.total_system_time: float = 0.0
        self.num_completed: int = 0
        self.num_abandoned: int = 0
        self.num_waited_over_break: int = 0
        # Для временной средней длины очередей: интеграл длины по времени
        self.queue_area: List[float] = [0.0 for _ in range(num_booths)]
        self.last_queue_time: float = 0.0

    def accumulate_queue_lengths(self, booths: List[TicketBooth], now: float):
        dt = now - self.last_queue_time
        if dt < 0:
            return
        for i, booth in enumerate(booths):
            self.queue_area[i] += len(booth.queue) * dt
        self.last_queue_time = now

    def record_service_completion(self, p: Passenger):
        if p.service_start_time is not None:
            self.total_wait_time += (p.service_start_time - p.arrival_time)
        if p.service_end_time is not None:
            self.total_system_time += (p.service_end_time - p.arrival_time)
        self.num_completed += 1

    def record_abandon(self):
        self.num_abandoned += 1

    def record_waited_over_break(self):
        self.num_waited_over_break += 1
```

```python
def expovariate(rate: float) -> float:
    # rate = 0 -> бесконечность, но в параметрах не используется
    u = random.random()
    return -math.log(1.0 - u) / rate

class BusStationModel:
    def __init__(self, params: Dict[str, Any]):
        self.params = params
        self.des = DES()
        self.booths: List[TicketBooth] = [TicketBooth(i, params) for i in range(params["num_booths"])]
        self.metrics = Metrics(params["num_booths"])
        self.now = 0.0
        self.next_passenger_id = 0

    # -------------------------
    # Инициализация
    # -------------------------
    def initialize(self):
        self.now = 0.0
        self.metrics.last_queue_time = 0.0
        # Планируем циклы для касс
        for booth in self.booths:
            announce_time, break_start, break_end = booth.plan_cycle(self.now)
            self.des.schedule(announce_time, kind="announce", payload={"booth": booth.booth_id})
            self.des.schedule(break_start, kind="break_start", payload={"booth": booth.booth_id})
            self.des.schedule(break_end, kind="break_end", payload={"booth": booth.booth_id})
        # Планируем первое прибытие пассажира
        ia = expovariate(self.params["arrival_rate"]) if self.params["arrival_rate"] > 0 else float("inf")
        self.des.schedule(self.now + ia, kind="arrival")

    # -------------------------
    # Выбор очереди (приблизительно минимальная)
    # -------------------------
    def choose_booth_for_passenger(self) -> int:
        delta = self.params["approx_delta"]
        epsilon = self.params["epsilon_random"]
        # Оцениваем воспринимаемую длину очереди
        perceived = [(booth.perceived_queue_length(), booth.booth_id) for booth in self.booths]
        perceived.sort()
        # лучший диапазон: все очереди чья длина <= min_len + delta
        min_len = perceived[0][0]
        candidates = [bid for (l, bid) in perceived if l <= min_len + delta]
        # с небольшой вероятностью возьмём любую кассу (модель нерациональности)
        if random.random() < epsilon:
            return random.choice([b.booth_id for b in self.booths])
        # иначе случайно среди лучших
        return random.choice(candidates)

    # -------------------------
    # Обработка событий
    # -------------------------
    def handle_event(self, evt: Event, now: float):
        self.metrics.accumulate_queue_lengths(self.booths, now)
        self.now = now
        kind = evt.kind
        payload = evt.payload or {}

        if kind == "arrival":
            self.on_arrival()
        elif kind == "service_end":
            self.on_service_end(payload["booth"], payload["passenger"]) 
        elif kind == "announce":
            self.on_announce(payload["booth"]) 
        elif kind == "break_start":
            self.on_break_start(payload["booth"]) 
        elif kind == "break_end":
            self.on_break_end(payload["booth"]) 
        else:
            pass

    def on_arrival(self):
        # Создаём пассажира
        p = Passenger(id=self.next_passenger_id, arrival_time=self.now)
        self.next_passenger_id += 1

        # Выбор кассы
        booth_id = self.choose_booth_for_passenger()
        p.chosen_booth_id = booth_id
        booth = self.booths[booth_id]

        # Проверка на отказ по лимиту ожидания (опционально)
        max_wait = self.params["max_wait_time"]
        if max_wait is not None and booth.is_on_break:
            # если касса на перерыве, ожидание как минимум до конца перерыва
            time_until_end = booth.next_cycle_break_end - self.now
            if time_until_end > max_wait:
                p.abandoned = True
                self.metrics.record_abandon()
                # не ставим в очередь
            else:
                booth.queue.append(p)
        else:
            booth.queue.append(p)

        # Если касса свободна и работает — запускаем обслуживание
        if not booth.is_on_break and booth.currently_serving is None and booth.queue:
            self.start_service(booth)

        # Планируем следующее прибытие
        ia = expovariate(self.params["arrival_rate"]) if self.params["arrival_rate"] > 0 else float("inf")
        next_t = self.now + ia
        if next_t <= self.params["simulation_duration"]:
            self.des.schedule(next_t, kind="arrival")

    def start_service(self, booth: TicketBooth):
        if booth.is_on_break or not booth.queue:
            return
        p = booth.queue.pop(0)
        booth.currently_serving = p
        p.service_start_time = self.now
        # Время обслуживания экспоненциальное
        service_time = expovariate(self.params["service_rate_per_booth"]) if self.params["service_rate_per_booth"] > 0 else float("inf")
        end_time = self.now + service_time
        # Если обслуживание пересекает перерыв, оно прерывается в момент break_start и продолжится после break_end
        if end_time > booth.next_cycle_break_start and not booth.is_on_break:
            # Считаем, что обслуживание прерывается на перерыв: остаток перенесём после перерыва
            time_before_break = booth.next_cycle_break_start - self.now
            remaining = service_time - max(0.0, time_before_break)
            # фиксируем, что человек будет ждать перерыв
            p.waited_over_break = True
            self.metrics.record_waited_over_break()
            # Планируем завершение после перерыва
            end_time = booth.next_cycle_break_end + remaining
        self.des.schedule(end_time, kind="service_end", payload={"booth": booth.booth_id, "passenger": p})

    def on_service_end(self, booth_id: int, p: Passenger):
        booth = self.booths[booth_id]
        # Завершение обслуживания
        booth.currently_serving = None
        p.service_end_time = self.now
        self.metrics.record_service_completion(p)
        # Запуск следующего, если возможно
        if not booth.is_on_break and booth.queue:
            self.start_service(booth)

    def on_announce(self, booth_id: int):
        booth = self.booths[booth_id]
        booth.on_announce()
        # Миграция пассажиров с вероятностью
        migrate_prob = self.params["announce_migrate_prob"]
        if migrate_prob <= 0 or not booth.queue:
            return
        # Пассажиры, не обслуживаемые (т.е. вся очередь) — могут мигрировать
        remaining_queue = []
        for p in booth.queue:
            if random.random() < migrate_prob:
                # выбрать новую кассу по тому же правилу
                new_booth_id = self.choose_booth_for_passenger()
                if new_booth_id != booth_id:
                    self.booths[new_booth_id].queue.append(p)
                    continue
            remaining_queue.append(p)
        booth.queue = remaining_queue
        # Если касса свободна и работает — запустить обслуживание, т.к. очередь изменилась
        if not booth.is_on_break and booth.currently_serving is None and booth.queue:
            self.start_service(booth)

    def on_break_start(self, booth_id: int):
        booth = self.booths[booth_id]
        booth.on_break_start()
        # Если кто-то обслуживался — это обслуживание прервётся в start_service логике при планировании
        # Остановить обслуживание: текущее завершение уже перенесено в будущий event, мы не отменяем событий
        # Новых запусков до конца перерыва не будет

    def on_break_end(self, booth_id: int):
        booth = self.booths[booth_id]
        booth.on_break_end(new_cycle_start=self.now)
        # Спланировать новый цикл
        announce_time, break_start, break_end = booth.plan_cycle(self.now)
        self.des.schedule(announce_time, kind="announce", payload={"booth": booth.booth_id})
        self.des.schedule(break_start, kind="break_start", payload={"booth": booth.booth_id})
        self.des.schedule(break_end, kind="break_end", payload={"booth": booth.booth_id})
        # Продолжить обслуживание, если очередь есть
        if booth.queue and booth.currently_serving is None:
            self.start_service(booth)

    # -------------------------
    # Запуск моделирования
    # -------------------------
    def run(self):
        self.initialize()
        self.des.run(until=self.params["simulation_duration"], on_event=self.handle_event)
        # Финальный апдейт площади очередей до конца моделирования
        self.metrics.accumulate_queue_lengths(self.booths, self.params["simulation_duration"])
```

```python
def compute_results(model: BusStationModel):
    m = model.metrics
    sim_time = model.params["simulation_duration"]
    avg_wait = (m.total_wait_time / m.num_completed) if m.num_completed > 0 else 0.0
    avg_system = (m.total_system_time / m.num_completed) if m.num_completed > 0 else 0.0
    avg_queue_lengths = [area / sim_time for area in m.queue_area]
    return {
        "avg_wait_time": avg_wait,
        "avg_system_time": avg_system,
        "avg_queue_lengths": avg_queue_lengths,
        "num_waited_over_break": m.num_waited_over_break,
        "num_completed": m.num_completed,
        "num_abandoned": m.num_abandoned,
    }

# Демонстрационный прогон
model = BusStationModel(PARAMS)
model.run()
results = compute_results(model)
results
```

```python
class MetricsTS(Metrics):
    def __init__(self, num_booths: int):
        super().__init__(num_booths)
        self.times: List[float] = []
        self.queues_history: List[List[int]] = []  # shape: [t][booth]
        self.status_history: List[List[int]] = []  # 1=work, 0=announce, -1=break
        self.system_size_history: List[int] = []

    def record_snapshot(self, booths: List[TicketBooth], now: float):
        self.times.append(now)
        self.queues_history.append([len(b.queue) for b in booths])
        statuses = []
        total = 0
        for b in booths:
            if b.is_on_break:
                statuses.append(-1)
            elif b.is_in_announce:
                statuses.append(0)
            else:
                statuses.append(1)
            total += len(b.queue) + (1 if b.currently_serving is not None else 0)
        self.status_history.append(statuses)
        self.system_size_history.append(total)

class BusStationModelLog(BusStationModel):
    def __init__(self, params: Dict[str, Any]):
        super().__init__(params)
        # Заменим метрики на версию с таймсериями
        self.metrics = MetricsTS(params["num_booths"])
        self.event_log: List[Dict[str, Any]] = []

    def log_event(self, kind: str, **kwargs):
        rec = {"time": self.now, "kind": kind}
        rec.update(kwargs)
        self.event_log.append(rec)

    def handle_event(self, evt: Event, now: float):
        # сначала как обычно накапливаем площадь очередей
        self.metrics.accumulate_queue_lengths(self.booths, now)
        self.now = now
        kind = evt.kind
        payload = evt.payload or {}
        if kind == "arrival":
            self.on_arrival()
            self.log_event("arrival")
        elif kind == "service_end":
            self.on_service_end(payload["booth"], payload["passenger"]) 
            self.log_event("service_end", booth=payload["booth"], passenger=payload["passenger"].id)
        elif kind == "announce":
            self.on_announce(payload["booth"]) 
            self.log_event("announce", booth=payload["booth"]) 
        elif kind == "break_start":
            self.on_break_start(payload["booth"]) 
            self.log_event("break_start", booth=payload["booth"]) 
        elif kind == "break_end":
            self.on_break_end(payload["booth"]) 
            self.log_event("break_end", booth=payload["booth"]) 
        # снимок состояния после обработки события
        casted: MetricsTS = self.metrics  # type: ignore
        casted.record_snapshot(self.booths, self.now)

    def on_arrival(self):
        # Полная реализация с логами (на основе базовой)
        p = Passenger(id=self.next_passenger_id, arrival_time=self.now)
        self.next_passenger_id += 1
        booth_id = self.choose_booth_for_passenger()
        p.chosen_booth_id = booth_id
        booth = self.booths[booth_id]
        max_wait = self.params["max_wait_time"]
        if max_wait is not None and booth.is_on_break:
            time_until_end = booth.next_cycle_break_end - self.now
            if time_until_end > max_wait:
                p.abandoned = True
                self.metrics.record_abandon()
                self.log_event("abandon", reason="break_wait", passenger=p.id, booth=booth_id)
            else:
                booth.queue.append(p)
                self.log_event("arrival_join", passenger=p.id, booth=booth_id)
        else:
            booth.queue.append(p)
            self.log_event("arrival_join", passenger=p.id, booth=booth_id)
        if not booth.is_on_break and booth.currently_serving is None and booth.queue:
            self.start_service(booth)
        ia = expovariate(self.params["arrival_rate"]) if self.params["arrival_rate"] > 0 else float("inf")
        next_t = self.now + ia
        if next_t <= self.params["simulation_duration"]:
            self.des.schedule(next_t, kind="arrival")

    def start_service(self, booth: TicketBooth):
        if booth.is_on_break or not booth.queue:
            return
        p = booth.queue.pop(0)
        booth.currently_serving = p
        p.service_start_time = self.now
        self.log_event("service_start", booth=booth.booth_id, passenger=p.id)
        service_time = expovariate(self.params["service_rate_per_booth"]) if self.params["service_rate_per_booth"] > 0 else float("inf")
        end_time = self.now + service_time
        if end_time > booth.next_cycle_break_start and not booth.is_on_break:
            time_before_break = booth.next_cycle_break_start - self.now
            remaining = service_time - max(0.0, time_before_break)
            p.waited_over_break = True
            self.metrics.record_waited_over_break()
            self.log_event("waited_over_break", booth=booth.booth_id, passenger=p.id)
            end_time = booth.next_cycle_break_end + remaining
        self.des.schedule(end_time, kind="service_end", payload={"booth": booth.booth_id, "passenger": p})

    def on_announce(self, booth_id: int):
        booth = self.booths[booth_id]
        booth.on_announce()
        migrate_prob = self.params["announce_migrate_prob"]
        if migrate_prob <= 0 or not booth.queue:
            return
        remaining_queue = []
        for p in booth.queue:
            if random.random() < migrate_prob:
                new_booth_id = self.choose_booth_for_passenger()
                if new_booth_id != booth_id:
                    self.booths[new_booth_id].queue.append(p)
                    self.log_event("migrate", passenger=p.id, src=booth_id, dst=new_booth_id)
                    continue
            remaining_queue.append(p)
        booth.queue = remaining_queue
        if not booth.is_on_break and booth.currently_serving is None and booth.queue:
            self.start_service(booth)

```

```python
# Функции визуализации

def plot_queue_lengths(model: BusStationModelLog):
    mts: MetricsTS = model.metrics  # type: ignore
    t = np.array(mts.times)
    if t.size == 0:
        print("Нет данных для графика")
        return
    fig, ax = plt.subplots(figsize=(10, 4))
    qs = np.array(mts.queues_history)
    for i in range(qs.shape[1]):
        ax.step(t, qs[:, i], where='post', label=f'Очередь кассы {i}')
    ax.set_xlabel('Время, мин')
    ax.set_ylabel('Длина очереди, чел')
    ax.set_title('Длины очередей во времени')
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.show()

def plot_booth_status(model: BusStationModelLog):
    mts: MetricsTS = model.metrics  # type: ignore
    t = np.array(mts.times)
    if t.size == 0:
        print("Нет данных для графика")
        return
    fig, ax = plt.subplots(figsize=(10, 2 + model.params['num_booths'] * 0.4))
    st = np.array(mts.status_history)
    for i in range(st.shape[1]):
        ax.step(t, st[:, i] + i * 3, where='post', label=f'Касса {i}')
    ax.set_xlabel('Время, мин')
    ax.set_yticks([i * 3 for i in range(st.shape[1])])
    ax.set_yticklabels([f'Касса {i}' for i in range(st.shape[1])])
    ax.set_title('Статусы касс: 1=работа, 0=анонс, -1=перерыв')
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.show()

EVENT_Y = {
    'arrival': 0.0,
    'arrival_join': 0.2,
    'abandon': -0.2,
    'service_start': 1.0,
    'service_end': 1.2,
    'announce': 2.0,
    'migrate': 2.2,
    'break_start': 3.0,
    'break_end': 3.2,
}

COLORMAP = {
    'arrival': 'tab:blue',
    'arrival_join': 'tab:blue',
    'abandon': 'tab:red',
    'service_start': 'tab:green',
    'service_end': 'tab:olive',
    'announce': 'tab:purple',
    'migrate': 'tab:orange',
    'break_start': 'tab:gray',
    'break_end': 'tab:gray',
}

MARKERS = {
    'arrival': 'o',
    'arrival_join': '^',
    'abandon': 'x',
    'service_start': 's',
    'service_end': 'd',
    'announce': 'P',
    'migrate': 'v',
    'break_start': '|',
    'break_end': '|',
}

def plot_event_timeline(model: BusStationModelLog):
    if not model.event_log:
        print('Журнал событий пуст')
        return
    fig, ax = plt.subplots(figsize=(12, 4))
    for ev in model.event_log:
        k = ev['kind']
        t = ev['time']
        y = EVENT_Y.get(k, 0.0)
        c = COLORMAP.get(k, 'k')
        m = MARKERS.get(k, '.')
        ax.scatter([t], [y], c=c, marker=m, s=20)
    ax.set_xlabel('Время, мин')
    ax.set_yticks(sorted(set(EVENT_Y.values())))
    reverse_map = {}
    for name, val in EVENT_Y.items():
        reverse_map.setdefault(val, []).append(name)
    ax.set_yticklabels(['/'.join(reverse_map[val]) for val in sorted(reverse_map)])
    ax.set_title('Лента событий (тип события во времени)')
    ax.grid(True, alpha=0.3)
    plt.show()

```

```python
# Дискретные отклики: бегущие средние по событиям service_end

def _extract_wait_system_series(model: BusStationModelLog):
    # Восстанавливаем по журналу событий времена прихода, старта и конца обслуживания
    arrivals: Dict[int, float] = {}
    starts: Dict[int, float] = {}
    ends: Dict[int, float] = {}
    for ev in model.event_log:
        k = ev['kind']
        t = ev['time']
        if k == 'arrival_join' and 'passenger' in ev:
            arrivals[ev['passenger']] = t
        elif k == 'service_start' and 'passenger' in ev:
            starts[ev['passenger']] = t
        elif k == 'service_end' and 'passenger' in ev:
            ends[ev['passenger']] = t
    # Формируем список завершённых с корректными метриками
    triplets = []
    for pid, t_end in ends.items():
        t_arr = arrivals.get(pid)
        t_start = starts.get(pid)
        if t_arr is None or t_start is None:
            continue
        wait = t_start - t_arr
        syst = t_end - t_arr
        triplets.append((t_end, wait, syst))
    triplets.sort(key=lambda x: x[0])
    # Бегущие средние
    times = []
    avg_wait = []
    avg_syst = []
    sum_w = 0.0
    sum_s = 0.0
    n = 0
    for t_end, w, s in triplets:
        n += 1
        sum_w += w
        sum_s += s
        times.append(t_end)
        avg_wait.append(sum_w / n)
        avg_syst.append(sum_s / n)
    return times, avg_wait, avg_syst


def plot_discrete_running_avgs(model: BusStationModelLog):
    times, avg_wait, avg_syst = _extract_wait_system_series(model)
    if not times:
        print('Нет завершённых обслуживаний для расчёта средних')
        return
    fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharex=True)
    ax[0].step(times, avg_wait, where='post', color='tab:blue')
    ax[0].set_title('Бегущее среднее ожидания в очереди')
    ax[0].set_xlabel('Время, мин')
    ax[0].set_ylabel('Минуты')
    ax[0].grid(True, alpha=0.3)

    ax[1].step(times, avg_syst, where='post', color='tab:green')
    ax[1].set_title('Бегущее среднее пребывания в системе')
    ax[1].set_xlabel('Время, мин')
    ax[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# Дискретные отклики по k: точки и кумулятивное среднее
def plot_discrete_running_avgs_by_count(model: BusStationModelLog):
    arrivals: Dict[int, float] = {}
    starts: Dict[int, float] = {}
    ends: Dict[int, float] = {}
    for ev in model.event_log:
        k = ev['kind']
        if k == 'arrival_join' and 'passenger' in ev:
            arrivals[ev['passenger']] = ev['time']
        elif k == 'service_start' and 'passenger' in ev:
            starts[ev['passenger']] = ev['time']
        elif k == 'service_end' and 'passenger' in ev:
            ends[ev['passenger']] = ev['time']

    records = []
    for pid, t_end in ends.items():
        t_arr = arrivals.get(pid)
        t_start = starts.get(pid)
        if t_arr is None or t_start is None:
            continue
        wait = t_start - t_arr
        syst = t_end - t_arr
        records.append((t_end, wait, syst))

    if not records:
        print('Нет завершённых обслуживаний для расчёта средних')
        return

    records.sort(key=lambda x: x[0])  # по времени завершения
    
    # Извлекаем отдельные значения времени ожидания и времени пребывания
    wait_times = [w for (_t, w, s) in records]
    system_times = [s for (_t, w, s) in records]
    
    # Проверка: выводим первые несколько значений для отладки
    if len(wait_times) > 0:
        print(f"Первые 5 значений wait_times: {wait_times[:5]}")
        print(f"Первые 5 значений system_times: {system_times[:5]}")
        print(f"Среднее wait_time: {np.mean(wait_times):.2f}, Среднее system_time: {np.mean(system_times):.2f}")
    
    # Вычисляем кумулятивные средние
    avg_wait = []
    avg_syst = []
    sw = 0.0
    ss = 0.0
    for i in range(len(wait_times)):
        sw += wait_times[i]
        ss += system_times[i]
        n = i + 1
        avg_wait.append(sw / n)
        avg_syst.append(ss / n)

    k_vals = np.arange(1, len(wait_times) + 1, dtype=int)
    wait_times = np.asarray(wait_times, dtype=float)
    system_times = np.asarray(system_times, dtype=float)
    avg_wait = np.asarray(avg_wait, dtype=float)
    avg_syst = np.asarray(avg_syst, dtype=float)

    # График 1: Среднее время ожидания пассажира в очереди
    fig1, ax1 = plt.subplots(figsize=(10, 6))
    ax1.scatter(k_vals, wait_times, color='blue', s=30, alpha=0.6, label='points')
    ax1.plot(k_vals, avg_wait, color='orange', linestyle='--', linewidth=2, label='cumulative avg')
    ax1.set_title('Wait (min)')
    ax1.set_xlabel('k')
    ax1.set_ylabel('Wait (min)')
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
    
    # График 2: Среднее время пребывания в системе
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    ax2.scatter(k_vals, system_times, color='blue', s=30, alpha=0.6, label='points')
    ax2.plot(k_vals, avg_syst, color='orange', linestyle='--', linewidth=2, label='cumulative avg')
    ax2.set_title('System Time (min)')
    ax2.set_xlabel('k')
    ax2.set_ylabel('System Time (min)')
    ax2.grid(True, alpha=0.3)
    ax2.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
 
```

```python
# Непрерывные отклики: бегущая средняя Lq(t) и кумулятивные количества

def plot_running_avg_queue_lengths(model: BusStationModelLog):
    mts: MetricsTS = model.metrics  # type: ignore
    t = np.array(mts.times)
    if t.size < 2:
        print('Недостаточно точек для расчёта Lq(t)')
        return
    qs = np.array(mts.queues_history)
    num_booths = qs.shape[1]
    areas = np.zeros(num_booths, dtype=float)
    running = np.zeros((t.size, num_booths), dtype=float)
    t0 = t[0]
    for i in range(1, t.size):
        dt = t[i] - t[i-1]
        areas += qs[i-1, :] * dt
        denom = max(t[i] - t0, 1e-9)
        running[i, :] = areas / denom
    fig, ax = plt.subplots(figsize=(10, 4))
    for b in range(num_booths):
        ax.step(t, running[:, b], where='post', label=f'Касса {b}')
    ax.set_xlabel('Время, мин')
    ax.set_ylabel('Средняя длина очереди')
    ax.set_title('Бегущая временная средняя длины очереди Lq(t)')
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.show()


def plot_cumulative_counts(model: BusStationModelLog):
    # Кумулятивы по событиям service_end (обслуженные) и waited_over_break (ждали перерыв)
    times_serv, times_wait_over = [], []
    for ev in model.event_log:
        if ev['kind'] == 'service_end':
            times_serv.append(ev['time'])
        elif ev['kind'] == 'waited_over_break':
            times_wait_over.append(ev['time'])
    times_serv.sort()
    times_wait_over.sort()

    def make_cum(times):
        if not times:
            return np.array([]), np.array([])
        xs, ys, c = [times[0]], [1], 1
        for t in times[1:]:
            xs.append(t); ys.append(c)
            c += 1
            xs.append(t); ys.append(c)
        return np.array(xs), np.array(ys)

    xs_s, ys_s = make_cum(times_serv)
    xs_w, ys_w = make_cum(times_wait_over)

    fig, ax = plt.subplots(figsize=(10, 4))
    if xs_s.size:
        ax.step(xs_s, ys_s, where='post', label='Обслуженные всего', color='tab:green')
    if xs_w.size:
        ax.step(xs_w, ys_w, where='post', label='Ждали перерыв всего', color='tab:orange')
    ax.set_xlabel('Время, мин')
    ax.set_ylabel('Кумулятивное количество, чел')
    ax.set_title('Кумулятивные количества: обслуженные и ждавшие перерыв')
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.show()
```

```python
# Много прогонов и агрегация

def run_replications(n: int, params: Dict[str, Any], seed: int = SEED) -> Tuple[list, list]:
    results = []
    models = []
    for i in range(n):
        random.seed(seed + i)
        m = BusStationModelLog(params)
        m.run()
        res = compute_results(m)
        results.append(res)
        models.append(m)
    return results, models

from statistics import mean

def aggregate_results(results: list) -> Dict[str, Any]:
    if not results:
        return {}
    keys = results[0].keys()
    agg = {}
    for k in keys:
        if isinstance(results[0][k], list):
            # усреднение поэлементно для списков
            L = len(results[0][k])
            agg[k] = [mean([r[k][j] for r in results]) for j in range(L)]
        else:
            agg[k] = mean([r[k] for r in results])
    return agg

# Демонстрация N прогонов
N = 10
results_N, models_N = run_replications(N, PARAMS, seed=SEED)
agg = aggregate_results(results_N)
print('Агрегированные метрики по', N, 'прогонам:')
for k, v in agg.items():
    print(k, ':', v)

# Построение графиков по первому прогону (пример)
model0: BusStationModelLog = models_N[0]
plot_queue_lengths(model0)
plot_booth_status(model0)
plot_event_timeline(model0)
# plot_discrete_running_avgs(model0)

plot_discrete_running_avgs_by_count(model0)

plot_running_avg_queue_lengths(model0)
plot_cumulative_counts(model0)


```
